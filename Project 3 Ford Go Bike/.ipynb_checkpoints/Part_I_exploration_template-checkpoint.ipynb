### Data Cleaning

```python
df_clean = df.copy()
```

```python
#Ensuring Proper Column data types
import datetime

dtype= {'start_time': 'datetime64', 
        'end_time': 'datetime64',
        'start_station_id': 'object',
        'end_station_id': 'object',
        'start_station_latitude': 'object',
        'start_station_longitude':  'object',
        'end_station_latitude':  'object',
        'end_station_longitude':  'object',
        'bike_id': 'object',
        'user_type': 'category'}
df_clean = df_clean.astype(dtype)
df_clean.info()
```

## Univariate Exploration

```python
# How are the users of the bike riding platform categorized?
sb.set_style("darkgrid")
category_order = ["Subscriber", "Customer"]
sb.catplot(x="user_type", kind="count", data = df_clean, order = category_order)
plt.title("User Category Distribution")
plt.xlabel("User Type")
plt.ylabel("Count") 
plt.show()
```

##### The countplot tells us that the majority of the users in San Francisco are Subscribers.

```python
df_clean.duration_sec.describe()
```

```python
# How long are the rides?
plt.figure(figsize=(10,6))
bins = np.arange(0, 10000, 50)

plt.hist(df_clean.duration_sec, bins=bins)
plt.title('Duration of the Rides');
```

##### Our distribution is skewed to the right and clustered around the mean of ~ 726 seconds or roughly 12 minutes.

```python
# What percent of users are subscribers and customers?
colors = sb.color_palette('pastel')[0:5]

plt.figure(figsize=(10,6))
sorted_counts = df_clean['user_type'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False, autopct='%1.2f%%', colors=colors);
plt.axis('square')
plt.title('Bike User Count Comparison', fontsize=15);
```

##### Around ~89% of users of the bikes are already subscribers.

### Since I want to investigate on date properties and it was not readily available in the dataset, we create those columns.

```python
df_clean['month'] = df_clean['start_time'].dt.month_name()
df_clean['start_day'] = df_clean['start_time'].dt.day_name()
df_clean['hour'] = df_clean['start_time'].dt.hour

df_clean
```

```python
# What is the peak hour for bicycle rides?
plt.style.use("default")
category_order = np.arange(0,24)
sb.catplot(x="hour", kind="count", data = df_clean)
plt.title("Bike Usage By Hour of the Day.")
plt.xlabel("Hour of the Day")
plt.ylabel("Count") 
plt.show()
```

##### Most of the bike rides are in the hours of 8-9 when people ae getting to work and 5-6 when they are leaving.

```python
df_clean["start_day"].value_counts()
```

```python
# What day of the week takes majority of the rides?
sb.set_style("darkgrid")
sb.set_palette("PRGn")
category_order = ["Thursday", "Tuesday", "Wednesday", "Friday", "Monday", "Sunday","Saturday"]
sb.catplot(x="start_day", kind="count", data = df_clean, order = category_order)
plt.title("Bike Usage By Day of Week.")
plt.xlabel("Day of the Week")
plt.xticks(rotation=45)
plt.ylabel("Count") 
plt.show()
```

##### Majority of the bike rides are On Thusday. Furthermore, most of the rides are during the week possibly because of people commuting to work.

```python
# What is the most Common Start Station?
plt.title('Most Popular Start Stations', fontsize=15)
df_clean.start_station_name.value_counts(ascending=True).tail(10).plot.barh();
```

##### Market St at 10th St is the most popular start station.

```python
# What is the most Common end Station?
plt.title('Most Popular end Stations', fontsize=15)
df_clean.end_station_name.value_counts(ascending=True).tail(10).plot.barh();
```

##### San Francisco Caltrain Station 2 (Townsend St at 4th St) is the most popular end station.

```python
# What is the least popular start station?
plt.title('Least Popular Start Stations', fontsize=15)
df_clean.start_station_name.value_counts(ascending=True).head(10).plot.barh();
```

##### 16th St Depot is the least popular start station.

```python
# What is the least popular end station?
plt.title('Least Popular End Stations', fontsize=15)
df_clean.end_station_name.value_counts(ascending=True).head(10).plot.barh();
```

##### Willow St at Vine St is the least popular end station.

## Bivariate Exploration

```python
#How do the categories compare by median, 25th and 75th percentiles? 
sb.catplot(x="user_type", y = "duration_sec", data = df_clean, kind = "box",sym="", whis=[5,95])
```

##### The customer segment had a higher median, 25th percentile and 75th percentile compared to the Subscriber segment.

```python
# How do the Categories compare by category during the week?
plt.title('Bike Usage by Day per Category', fontsize=15)
g = sb.countplot(data=df_clean, x='start_day', hue='user_type')
g.set(xlabel='Month', ylabel='')
sb.despine(fig=None, ax=None, top=True, right=True, left=True, bottom=False, offset=None, trim=False)
plt.gca().legend().set_title('');
```

##### Subscibers had a higher usage during the week than customers.

```python
# How does the median of time spent riding compare by Gender?
from statistics import median
sb.catplot(x="member_gender", y="duration_sec", data = df_clean, kind="point", estimator=median,join=False,capsize=0.2)
```

### Talk about some of the relationships you observed in this part of the investigation. How did the feature(s) of interest vary with other features in the dataset?

> There is a of correlation between duration and user type in regard to the fact that Subscribers tends to have a shorter usage time than Customers.

### Did you observe any interesting relationships between the other features (not the main feature(s) of interest)?

> Subscriber usage time is peak between 8 and 9 as well as 5 and 6 contrary to Customers with a more flexible usage time.

## Multivariate Exploration

```python
plt.figure(figsize=(10,6))
plt.title('Hourly Duration Usage', fontsize=15)
sb.barplot(data=df_clean, x='hour', y='duration_sec', ci=None, hue='user_type')
sb.despine(fig=None, ax=None, top=True, right=True, left=True, bottom=False, offset=None, trim=False)
plt.gca().legend().set_title('');
```

```python
fig = plt.figure(figsize = [8,6])
ax = sb.pointplot(data = df_clean, x ='start_day', y = 'duration_sec', hue = 'user_type',join=None)
plt.title('Trip Duration across week days and customer type')
plt.ylabel('Average Trip Duration (Secs)')
ax.set_yticklabels([],minor = True)
plt.show();
```

```python
sb.pointplot(data = df_clean, x ='member_gender', y = 'duration_sec', hue = 'bike_share_for_all_trip',join=None)
plt.title('Trip Duration across gender and bike_share_for_all_trip')
plt.ylabel('Average Trip Duration (Secs)')
ax.set_yticklabels([],minor = True)
plt.show();
```

### Talk about some of the relationships you observed in this part of the investigation. Were there features that strengthened each other in terms of looking at your feature(s) of interest?

> Subcribers have a more set usage period while Customers have a flexible time for riding. 

### Were there any interesting or surprising interactions between features?

> It is surprising that customers have a peak usage time at 3: 00 am.
